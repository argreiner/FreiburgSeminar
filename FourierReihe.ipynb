```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = np.linspace(0.,1.0,1000)
omega = 2.0*np.pi*np.arange(100)
ampsin = np.random.rand(100)
ampcos = np.random.rand(100)
y = np.zeros((1000))
```

```python
%matplotlib notebook

for i in np.arange(100):
    y = y + ampcos[i]*np.cos(omega[i]*T) + ampsin[i]*np.sin(omega[i]*T)  # y+= np.sin(omega[i]*T)
#
plt.plot(T,y)
```

```python
%matplotlib notebook
y=np.random.rand(1000)
f = np.fft.fft(y)
plt.plot(np.real(f))
plt.plot(y*50)
```

```python
%matplotlib notebook
plt.plot(np.fft.ifft(f)-y)
```
