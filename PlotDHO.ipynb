```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
fig, (ax1,ax2) = plt.subplots(1,2) # Let's make two subplot side by side addressed by ax1 ans ax2
#
t=np.arange(0,300,0.001)
x0=1.
v0=0.
omega=1.5
gamma=1.0
omega0=1.0
omegar=np.sqrt(4.*omega0**2-gamma**2/4)
factor=1/(omega**2-1)
x=np.exp(-gamma*t/2.)*(np.cos(omegar*t)*x0+np.sin(omegar*t)*v0)-factor*np.cos(omega*t)+factor
v=-np.exp(-gamma*t/2.)*(np.sin(omegar*t)*x0+np.cos(omegar*t)*v0)+omega*factor*np.sin(omega*t)
#
ax1.set_title('Phase space v(t) over x(t)') # This sets the title of the left plot
ax1.set_aspect(1)       # set the aspect ratio of the graph
ax1.set_xlabel('x(t)')  # the label on the x-axes
ax1.set_ylabel('v(t)')  # the label on the y-axes
ax1.plot(x,v,'-')       # phase space plot, i.e. y(x)
#
ax2.set_aspect(100)      # set the apsect ...
ax2.set_xlabel('Time t')
ax2.set_ylabel('Amplitude')
ax2.yaxis.tick_right()
ax2.yaxis.set_label_position("right")
ax2.plot(t[10000:],x[10000:],'r')       # plot x(t)
ax2.plot(t[10000:],v[10000:],'b')       # plot y(t)
ax2.set_title('Amplitude x(t) and y(t)')
fig.savefig('DHOSystem.svg', format = 'svg', dpi=300) # save the figure as a vector graphic
fig.savefig('DHOSystem.pdf', format = 'pdf') # save it as a pdf
fig.suptitle('Predator prey system')
```
